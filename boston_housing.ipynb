### Implementation: Define a Performance Metric
It is difficult to measure the quality of a given model without quantifying its performance over training and testing. This is typically done using some type of performance metric, whether it is through calculating some type of error, the goodness of fit, or some other useful measurement. For this project, you will be calculating the [*coefficient of determination*](http://stattrek.com/statistics/dictionary.aspx?definition=coefficient_of_determination), R<sup>2</sup>, to quantify your model's performance. The coefficient of determination for a model is a useful statistic in regression analysis, as it often describes how "good" that model is at making predictions. 

The values for R<sup>2</sup> range from 0 to 1, which captures the percentage of squared correlation between the predicted and actual values of the **target variable**. A model with an R<sup>2</sup> of 0 always fails to predict the target variable, whereas a model with an R<sup>2</sup> of 1 perfectly predicts the target variable. Any value between 0 and 1 indicates what percentage of the target variable, using this model, can be explained by the **features**. *A model can be given a negative R<sup>2</sup> as well, which indicates that the model is no better than one that naively predicts the mean of the target variable.*

For the `performance_metric` function in the code cell below, you will need to implement the following:
- Use `r2_score` from `sklearn.metrics` to perform a performance calculation between `y_true` and `y_predict`.
- Assign the performance score to the `score` variable.

```python
# TODO: Import 'r2_score'
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    
    # TODO: Calculate the performance score between 'y_true' and 'y_predict'
    score = r2_score(y_true, y_predict)
    
    # Return the score
    return score
```

### Question 3 - Training and Testing
*What is the benefit to splitting a dataset into some ratio of training and testing subsets for a learning algorithm?*  
**Hint:** What could go wrong with not having a way to test your model?

**Answer: **

Splitting a dataset into separate training and testing subsets is best understood in the context of the [bias-variance tradeoff](http://scott.fortmann-roe.com/docs/BiasVariance.html) (see Figure 6 in Section 4.4).  More specifically, _without_ splitting a dataset into separate training and testing subsets, we have no way of estimating where on this tradeoff curve our model is performing, even though our goal is to train a classifier that falls along the "Optimum Model Complexity" performance line.

Whenever we are training a classifier, it could easily be that we train a classifier which is overly-attentive to irrelevant variations in the training data, and thus overfits the data, and thus does not generalize well to information it hasn't seen before.  Such a classifier would have a high accuracy on predicting the training data, but would likely have a poor accuracy on predicting new data.  Without a test set, however, we would only know that the classifier has high accuracy on the training set.

Similarly, it could just as easily be that we train a classifier which does not make full use of the dimensionality of the data, and thus could achieve a higher accuracy even without overfitting.  Or, it could be that our classifier is fundamentally ill-suited to fit the data it's training on (e.g. a linear regression to fit quadratic data, without a transformation into linear space).  In either case, such a model would be called "biased".  Without a test set, however, we would only know that the classifier does not have high accuracy on the test set.

The significance of the use of a test set is that it allows us to draw conclusions about which of the above error cases we find our model exists in.  Without that test set, there is no way for us to reason meaningfully about how our classifier performs on data it wasn't allowed to train on (often worded as: "data it's never seen before"), and that state of ambiguity strongly undermines confidence in the classifier.

With a test set, however, we have an indication about where on the bias-variance tradeoff curve we find ourselves, and therefore we also have suggestions about how to re-train the classifier to achieve greater predictive accuracy.
